### Схема таблиц и описание данных

<img src="https://pictures.s3.yandex.net/resources/photo_2019-11-08_14-08-31_1573733426.jpg" width=700 height=1100 />

База данных об авиаперевозках:

Таблица **`airports`** — информация об *аэропортах*:
* `airport_code` — трёхбуквенный код аэропорта,
* `airport_name` — название аэропорта,
* `city` — город,
* `timezone` — часовой пояс.

Таблица **`aircrafts`** — информация о *самолётах*:
* `aircraft_code` — код модели самолёта,
* `model` — модель самолёта,
* `range` — дальность полётов.

Таблица **`tickets`** — информация о *билетах*:
* `ticket_no` — уникальный номер билета,
* `passenger_id` — уникальный идентификатор пассажира,
* `passenger_name` — имя и фамилия пассажира.

Таблица **`flights`** — информация о *рейсах*:
* `flight_id` — уникальный идентификатор рейса,
* `departure_airport` — аэропорт вылета,
* `departure_time` — дата и время вылета,
* `arrival_airport` — аэропорт прилёта,
* `arrival_time` — дата и время прилёта,
* `aircraft_code` — уникальный идентификатор самолёта.

Таблица **`ticket_flights`** — стыковая таблица *«рейсы-билеты»*:
* `ticket_no` — номер билета,
* `flight_id` — уникальный идентификатор рейса.

Таблица **`festivals`** — информация о *фестивалях*:
* `festival_id` — уникальный номер фестиваля,
* `festival_date` — дата проведения фестиваля,
* `festival_city` — город проведения фестиваля,
* `festival_name` — название фестиваля.

Сохраним в перемнную `top10_average_flights` топ-10 городов по количеству рейсов:

```python
# сохранение в переменную 'top10_average_flights' топ-10 городов по количеству рейсов 
# методом nlargest() с удалением старых индексов методом reset_index(drop=True)
top10_average_flights = city_average_flights.nlargest(10, 'average_flights').reset_index(drop=True)

# вывод таблицы top10_average_flights
top10_average_flights
```

Для построения столбчатых диаграмм создадим функцию `plot_barplot()`, принимающую на вход датафрейм, столбцы по оси абсцисс и ординат, их подписи, а также название графика:

```python
# создание функции plot_barplot() с аргументами (data, x, x_label, y, y_label, title)
def plot_barplot(data, x, x_label, y, y_label, title):
    '''
    Возвращает столбчатую диаграмму таблицы 'data' по оси абсцисс 'x' и 
    ординат - 'y' с подписью соответствующих осей 'x_label', 'y_label' 
    и названием графика 'title'
    '''
    # построение столбчатой диаграммы таблицы 'data' по оси абсцисс 'x'
    # и ординат 'y', с подписью соответствующих осей 'x_label', 'y_label' 
    sns.barplot(x=x, y=y, data=data,\
                order=data.sort_values(x, ascending = False)[y]).\
                set(xlabel=x_label, ylabel=y_label)
    
    # вывод подписи графика
    plt.title(title, fontsize=15)

    # активация второстепенных делений осей
    plt.minorticks_on()
    # отображение основной сетки grid() 'major' и определение внешнего вида линий
    plt.grid(which='major', color = 'k', linewidth = 0.3)
    # отображение второстепенной сетки grid() 'minor' определение внешнего вида линий
    plt.grid(which='minor', color = 'k', linewidth = 0.1, linestyle = ':')
    # отображение графика
    plt.show
```

С помощью данной функции построим *столбчатую диаграмму количества рейсов с вылетом в сентябре 2018 года по моделям самолетов*:

```python
# установление размера поля figsize=(12, 6)
plt.figure(figsize=(12, 6))

# вызов функции plot_barplot() с соответствующими аргументами
plot_barplot(model_flights_amount, 'flights_amount', 'Количество рейсов ', 'model', 'Модели самолетов',\
             'Столбчатая диаграмма количества рейсов с вылетом в сентябре 2018 года по моделям самолетов')
```

* лидеры авиаперелетов: **Москва** - в связи с популярностью столицы и прибывающими рейсами для пересадки на международные авиалинии, **Санкт-Петербург**, являющийся популярным туристическим и деловым направлением для поездок и **Новосибирск** ввиду "удачного" расположения в центре страны, позволяющего делать пересадки на региональные авиалинии
